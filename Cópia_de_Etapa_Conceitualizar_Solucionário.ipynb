## **Atividade de experimentao 6**

<p align="justify">
Imagine que voc est trabalhando em um projeto de fsica e precisa calcular a velocidade final de um objeto em movimento. Voc sabe a velocidade inicial, a acelerao constante do objeto, e a distncia percorrida. Para isso escreva um programa em Python que pea ao usurio para inserir a velocidade inicial (em metros por segundo), a acelerao constante (em metros por segundo ao quadrado), e a distncia percorrida (em metros).
<p>

$$V^2= V_0^{2}+2a{\Delta}s$$


<p align="justify">
Calcule e exiba a velocidade final do objeto.
<p>

```python
print("Esse programa tem o objetivo de calcular a equao de Torricelli")
print("V = Vo + 2ads")
print("A fim de encontrar as velocidade final fornea os parametros iniciais.")

Vo = float(input("Insira da velocidade inicial em m/s: "))
a = float(input("Insira da aceleracao do objeto em m/s: "))
ds = float(input("Insira do deslocamento do objeto em m: "))

V = ((Vo**2)+(2*a*ds))**(0.5)

print("A velocidade final do mvel : ", V, "m/s")
```

## **Atividade de experimentao 19**

<p align="justify">
Escreva um programa em Python que simule o sistema de senha dos bancos, permitindo no mximo 3 tentativas antes de bloquear a senha. Se a senha estiver correta dentro das 3 tentativas, o programa deve permitir o acesso ao banco.
<p>

```python
senha_correta = "123456"  # Defina a senha correta
tentativas = 0
while tentativas < 3:
  senha = input("Digite a senha: ")
  if senha == senha_correta:
    print("Acesso permitido. Bem-vindo ao banco.")
    break  # Encerra o loop se a senha estiver correta
  else:
    tentativas += 1
    tentativas_restantes = 3 - tentativas
    if tentativas_restantes > 0:
      print(f"Senha incorreta. Voc tem {tentativas_restantes} tentativa(s) restante(s).")
    else:
      print("Senha incorreta. Acesso bloqueado. Entre em contato com o banco.")
```

## **Atividade de experimentao 37**

<p align="justify">
Crie e manipule um banco de dados SQLite utilizando Python, no qual contenha as informaes dos alunos do curso, como: id, nome e nota. Crie e conecte ao banco de dados Replit/SQLite, nomeando como alunosdocurso. Aborde e utilize as funes CRUD para tratar os dados.
<p>

```python
import sqlite3

# Conexo com o banco de dados
conn = sqlite3.connect('alunos_phyton.db')

# Criar uma tabela
cursor = conn.cursor()

cursor.execute('CREATE TABLE IF NOT EXISTS alunos (id INTEGER PRIMARY KEY, nome TEXT, nota REAL)')
```

```python
# Inserir dados na tabela = CREATE
cursor.execute('INSERT INTO alunos (id, nome, nota) VALUES (1, "Leandro", 8.5)')
cursor.execute('INSERT INTO alunos (id, nome, nota) VALUES (2, "Laise", 9.0)')
cursor.execute('INSERT INTO alunos (id, nome, nota) VALUES (3, "Michele", 7.0)')
```

```python
# Visualizar os dados da tabela = READ
cursor.execute('SELECT * FROM alunos')
alunos = cursor.fetchall()
print(alunos)
```

```python
# Atualizar os dados da tabela = UPDATE
cursor.execute('UPDATE alunos SET nota = 9.5 WHERE id = 1')
cursor.execute('UPDATE alunos SET nome = "Emilene" WHERE id = 1')

cursor.execute('SELECT * FROM alunos')
alunos = cursor.fetchall()
print(alunos)
```

```python
# Apagar dados da tabela = DELETE
cursor.execute('DELETE FROM alunos WHERE id = 2')

cursor.execute('SELECT * FROM alunos')
alunos = cursor.fetchall()
print(alunos)
```
